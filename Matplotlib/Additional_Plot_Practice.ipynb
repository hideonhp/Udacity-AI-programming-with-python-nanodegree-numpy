```python
# prerequisite package imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

from solutions_biv import additionalplot_solution_1, additionalplot_solution_2
```

We'll continue to make use of the fuel economy dataset in this workspace.

```python
fuel_econ = pd.read_csv('./data/fuel_econ.csv')
fuel_econ.head()
```

### **Task 1**: 
Plot the distribution of combined fuel mileage (column 'comb', in miles per gallon) by manufacturer (column 'make'), for all manufacturers with at least eighty cars in the dataset. Consider which manufacturer order will convey the most information when constructing your final plot. 

**Hint**: Completing this exercise will take multiple steps! Add additional code cells as needed in order to achieve the goal.

```python
# Filter the manufacturers with at least eighty cars
manufacturer_counts = fuel_econ['make'].value_counts()
large_manufacturers = manufacturer_counts.index[manufacturer_counts >= 80]

# Filter the dataset to include only cars from large manufacturers
fuel_econ_large = fuel_econ[fuel_econ['make'].isin(large_manufacturers)]

# Plot the distribution of combined fuel mileage by manufacturer
plt.figure(figsize=[12, 6])
sb.boxplot(data=fuel_econ_large, x='make', y='comb', color=sb.color_palette()[0])
plt.xticks(rotation=90)
plt.xlabel('Manufacturer')
plt.ylabel('Combined Fuel Mileage (mpg)')
plt.title('Distribution of Combined Fuel Mileage by Manufacturer (>=80 cars)')
plt.show()
```

### **Task 2**: 
Continuing on from the previous task, plot the mean fuel efficiency for each manufacturer with at least 80 cars in the dataset.

```python
# Calculate the mean fuel efficiency for each manufacturer
mean_fuel_efficiency = fuel_econ_large.groupby('make')['comb'].mean().sort_values()

# Plot the mean fuel efficiency
plt.figure(figsize=[12, 6])
sb.barplot(x=mean_fuel_efficiency.index, y=mean_fuel_efficiency, color=sb.color_palette()[0])
plt.xticks(rotation=90)
plt.xlabel('Manufacturer')
plt.ylabel('Mean Combined Fuel Mileage (mpg)')
plt.title('Mean Combined Fuel Mileage by Manufacturer (>=80 cars)')
plt.show()
```
